```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
import matplotlib.pyplot as plt

# 1) Charger les donnes depuis le fichier Excel
data = pd.read_excel('zzz02.xlsx')

# 2) Dfinir l'entre et la sortie
X = data.iloc[:, 0].values.reshape(-1, 1)
y1 = data.iloc[:, 1].values

# 3) Split train/test
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y1, test_size=0.2, random_state=42)

# 4) Crer et entraner le modle
regr_1 = DecisionTreeRegressor(max_depth=14)
regr_1.fit(X_train1, y_train1)

# 5) Prdiction
y_1 = regr_1.predict(X_test1)

# 6) valuation
mae = metrics.mean_absolute_error(y_test1, y_1)
mse = metrics.mean_squared_error(y_test1, y_1)
rmse = np.sqrt(mse)
r2 = metrics.explained_variance_score(y_test1, y_1)

# 7) Conversion en pourcentage (par rapport  la valeur max des donnes test)
Y_max = np.max(y_test1)
mae_percent = (mae / Y_max) * 100
rmse_percent = (rmse / Y_max) * 100
mse_percent = (mse / (Y_max**2)) * 100

# 8) Affichage
print("\n=== Erreurs brutes ===")
print("MAE :", mae)
print("MSE :", mse)
print("RMSE:", rmse)
print("R^2 :", r2)

print("\n=== Erreurs en pourcentage ===")
print("MAE (%)  :", mae_percent)
print("MSE (%)  :", mse_percent)
print("RMSE (%) :", rmse_percent)

# 7)  TRACAGE DE LA COURBE 

# Cration d'une figure de taille spcifique
plt.figure(figsize=(90, 80))

# Taille des points
point_size = 2500  # Vous pouvez ajuster cette valeur selon vos besoins

# Scatter plot for 0.10
plt.scatter(X_test1, y_1, color="red", label=" Predicted plot ", s=point_size)
plt.plot(X, y1, color="black", label=" Original plot", linewidth=10)


            

# Dfinir les tiquettes et la lgende
plt.xlabel("Time (Hours)", fontsize=150)  # Dfinir la taille de police pour l'axe x
plt.ylabel("Vehicle traffic (Number of Active vehicles) ", fontsize=150)  # Dfinir la taille de police pour l'axe y

# Dfinir les limites de l'axe x
#plt.xlim(1.6, 2.5)
#plt.ylim( -0.2, 4.8)
# Dfinir les tiquettes et la lgende


plt.legend(fontsize=100)  # Dfinir la taille de police pour la lgende

#plt.title("Experimental vs Predicted Data", fontsize=70)

# Agrandir la taille de la police des tiquettes des graduations des axes
plt.tick_params(axis='both', which='major', labelsize=100)


# paissir les lignes des axes
ax = plt.gca()  # Get current axes
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(10)  # paisseur de 10 points

# Afficher une grille
plt.grid(True)

# Afficher le plot
plt.show()

#=== Erreurs brutes ===
#MAE : 399.3636363636364
#MSE : 329276.36363636365
#RMSE: 573.8260743782594
#R^2 : 0.9877818524450335

#=== Erreurs en pourcentage ===
#MAE (%)  : 2.354460773279309
#MSE (%)  : 0.11444753396289754
#RMSE (%) : 3.383009517617376


 

```
