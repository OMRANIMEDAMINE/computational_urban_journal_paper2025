```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
import matplotlib.pyplot as plt

# ------------------- 1) Charger les donnÃ©es -------------------
data = pd.read_excel('zzz02.xlsx')

# EntrÃ©e = premiÃ¨re colonne
X = data.iloc[:, 0].values.reshape(-1, 1)

# Sortie = deuxiÃ¨me colonne
Y = data.iloc[:, 1].values

# ------------------- 2) Normalisation de X -------------------
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# ------------------- 3) Train/Test split -------------------
X_train, X_test, Y_train, Y_test = train_test_split(
    X_scaled, Y, test_size=0.2, random_state=42
)

# ------------------- 4) ModÃ¨le KNN -------------------
knn = KNeighborsRegressor(n_neighbors=7)
knn.fit(X_train, Y_train)

# ------------------- 5) PrÃ©diction sur test -------------------
Y_pred = knn.predict(X_test)

# ------------------- 6) Ãvaluation -------------------
mae = metrics.mean_absolute_error(Y_test, Y_pred)
mse = metrics.mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r2 = metrics.r2_score(Y_test, Y_pred)  # RÂ² classique

# Affichage des rÃ©sultats
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("RÂ² Score:", r2)

#Mean Absolute Error (MAE): 208.21428571428575
#Mean Squared Error (MSE): 114113.44155844161
#Root Mean Squared Error (RMSE): 337.80681100066886
#RÂ² Score: 0.9957591536080775

# --- TRACÃ CORRIGÃ ET AMÃLIORÃ ---
plt.figure(figsize=(90, 80))

# PrÃ©dictions (points rouges trÃ¨s visibles)
plt.scatter(X_test, Y_pred, color="red", label="Predicted Plot", s=2900)

# DonnÃ©es rÃ©elles (courbe noire + Ã©paissie)
plt.plot(X_scaled, Y, color="black", label="Original Plot", linewidth=10)

# Labels trÃ¨s grands
plt.xlabel("Time (Hours)", fontsize=150)
plt.ylabel("Vehicle Traffic (Number of Active Vehicles)", fontsize=150)

# Taille de la lÃ©gende
plt.legend(fontsize=120)

# Taille des marqueurs de graduation
plt.tick_params(axis='both', which='major', labelsize=120)

# Ãpaissir les axes
ax = plt.gca()
for axis in ['top', 'bottom', 'left', 'right']:
    ax.spines[axis].set_linewidth(10)

# Grille visible
plt.grid(True)

# Afficher le graphique
plt.show()
```
